```
import pandas as pd
import utils
import matplotlib.pyplot as plt
from math import sin, cos, sqrt, atan2 ,radians
from datetime import datetime
from datetime import timedelta 
#from matplotlib import pyplot as plt
%matplotlib inline
```

```
ds1 = utils.parse_gpx('data/ride1.gpx')
df1=ds1                     #storing in another data frame
df1=df1.drop_duplicates()  #Removing the common values
```

```
#Function for calculating Distance
def calculateDistance(lat1, lon1, lat2, lon2):
    R = 6373.0
    dlon =radians(lon2 - lon1)
    dlat =radians(lat2 - lat1)
    lat1 =radians(lat1)
    lat2 =radians(lat2)
    a = (sin(dlat/2))**2 + cos(lat1) * cos(lat2) * (sin(dlon/2))**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return distance #Returns Distance in KM

#Function for calculating Gradient
def calculateGradient(elev1,elev2):
    gradient=elev2-elev1
    return gradient

#Function for calculating Time
def calculateTime(Time1,Time2):
    Time=Time2-Time1
    Time=Time.total_seconds() #Convert the deltatime to seconds
    return Time/3600          #Convert to Hours,return time in Seconds
 
#Function for Calculating Speed    
def calculateSpeed(Distance1,Time1):
    Speed =Distance1/Time1
    return Speed  #returns Speed in KM/HR

distance =[]
Gradient =[]
Time =[]
#Time1=[]
Speed =[]
for i in range(len(df1)-1):
   
   #Store the distance calculated in List
   distance.append(calculateDistance(df1['latitude'][i], df1['longitude'][i], df1['latitude'][i+1], df1['longitude'][i+1]))
 
   #Store the Gradient calculated in List
   Gradient.append(calculateGradient(df1['elevation'][i],df1['elevation'][i+1]))
  
   #Store the Time calculated in List
   Time.append(calculateTime(df1.index[i],df1.index[i+1]))
   
  # df1['f'][i]=df1.index[i]
    
for j in range(len(distance)):
     #Store the speed calculated in List
    Speed.append(calculateSpeed(distance[j],Time[j]))
    
#Convert all the List into Series
A=pd.Series(Speed)
B=pd.Series(distance)
C=pd.Series(Time)
D=pd.Series(Gradient)
#D=pd.Series(Time1)

#Creatiing a New Dataframe which stores the calculated[Speed,Distance,Time]
df3=pd.DataFrame({'speed': A, 'Distance': B,'Time':C,'Gradient':D})

#Average
Avg=((df3['Distance'].sum())/(df3['Time'].sum()))

print("Total Distance Covered:",df3['Distance'].sum(),"KM")
print("Total Time Covered:",df3['Time'].sum(),"Hrs.")
print("Average Speed of the Rider is :",Avg,"KM/hr")
```
